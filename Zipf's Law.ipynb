```python
import matplotlib.pyplot as plt
import string
fp= open('Alice.txt', 'r', encoding='UTF-8')
line=fp.readline()
dic={}
while line:
    s=line.split()
    print(line)
    line=fp.readline()
    for w in s:
        w.translate(string.punctuation)
        w=w.translate(str.maketrans('','',string.punctuation))
        w=w.lower()
        if w not in dic:
            dic[w]=1
         
        else:
            dic[w]=dic[w]+1
print(dic)
listk=[]
listv=[]
for key in dic:
    listk.append(dic[key])
    
for value in dic:
    listv.append(dic[value])
listk.sort(reverse=True)
plt.loglog(range(len(listk)),listk)
plt.xlabel('Rank of the word')
plt.ylabel('frequency')
plt.plot(listk)
plt.show()
   
fp.close() 
```
